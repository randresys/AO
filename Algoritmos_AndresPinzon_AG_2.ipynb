```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
f = lambda X: X[0]**2 +X[1]**2
df = lambda X: [2*X[0],2*X[1]]
```

```python
resolucion = 100
rango = 2.5
X=np.linspace(-rango, rango, resolucion)
Y=np.linspace(-rango, rango, resolucion)
Z=np.zeros((resolucion,resolucion))
for ix,x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[iy,ix] = f([x,y])

contour = plt.contourf(X,Y,Z, resolucion)
cbar = plt.colorbar()

P = [random.uniform(-2,2 ), random.uniform(-2,2 )]
plt.plot(P[0], P[1],"o", c="red")
TA = 0.05
for _ in range(500):
  grad = df(P)
  TA = abs(np.mean(grad))/10

  P[0],P[1] = P[0] - TA*grad[0],P[1] - TA*grad[1]
  plt.plot(P[0], P[1],"o", c="red")

plt.plot(P[0], P[1],"o", c="white")
print("Solucion: ", P, f(P))
```

```python
f = lambda X: np.sin(1/2 * X[0]**2 - 1/4 * X[1]**2+3) * np.cos(2 * X[0] + 1 - np.e**X[1])

df = lambda X:[X[0] * np.cos(2*X[0]-np.e**X[1]+1) * np.cos(1/2*X[0]**2-1/4*X[1]**2+3)- 2*np.sin(2*X[0]-np.e**X[1]+1) * np.sin(1/2 * X[0]**2 - 1/4 * X[1]**2+3),
               np.e**X[1] * np.sin(2*X[0]-np.e**X[1]+1) * np.sin(1/2 * X[0]**2 - 1/4 * X[1]**2+3)-1/2*X[1] * np.cos(2*X[0]-np.e**X[1]+1) * np.cos(1/2*X[0]**2-1/4*X[1]**2+3)]
```

```python
resolucion = 100
rango = 5
X=np.linspace(-rango, rango, resolucion)
Y=np.linspace(-rango, rango, resolucion)
Z=np.zeros((resolucion,resolucion))
for ix,x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[iy,ix] = f([x,y])

contour = plt.contourf(X,Y,Z, resolucion)
cbar = plt.colorbar()

P = [random.uniform(-5,5 ), random.uniform(-5,5 )]
plt.plot(P[0], P[1],"o", c="red")
TA = 0.07
for _ in range(200):
  grad = df(P)
  P[0],P[1] = P[0] - TA*grad[0],P[1] - TA*grad[1]
  plt.plot(P[0], P[1],"o", c="red")

plt.plot(P[0], P[1],"o", c="white")
print("Solucion: ", P, f(P))
```
